```python
import zenoh
import heros
from heros import RemoteHERO
from time import sleep
import zmq
tmin = 30e-3

session = zenoh.open(zenoh.Config.from_file('zenoh_config.json5'))
session_mgr = heros.zenoh.session_manager
session_mgr._session = session
```

```python
# START MONITOR
session = zenoh.open(zenoh.Config.from_file('zenoh_config.json5'))
session_mgr = heros.zenoh.session_manager
session_mgr._session = session
with RemoteHERO('SCTL',session_manager=session_mgr) as Lock:
    Lock.start_monitor('Mon')
```

```python
# START ERROR MONITOR
session = zenoh.open(zenoh.Config.from_file('zenoh_config.json5'))
session_mgr = heros.zenoh.session_manager
session_mgr._session = session
with RemoteHERO('SCTL',session_manager=session_mgr) as Lock:
    Lock.start_error_monitor('Mon',tmin=tmin,zmq_pub=True) #Start error monitor with zmq enabled
```

```python
# START CAV LOCK
session = zenoh.open(zenoh.Config.from_file('zenoh_config.json5'))
session_mgr = heros.zenoh.session_manager
session_mgr._session = session
with RemoteHERO('SCTL',session_manager=session_mgr) as Lock:
    Lock.stop_loop('Cav')
    sleep(1)
    Lock.start_lock('Cav')
```

```python
# STOP CAV LOCK
session = zenoh.open(zenoh.Config.from_file('zenoh_config.json5'))
session_mgr = heros.zenoh.session_manager
session_mgr._session = session
with RemoteHERO('SCTL',session_manager=session_mgr) as Lock:
    print('opened')
    Lock.stop_loop('Cav')
    sleep(0.5)
    Lock.start_scan('Cav')
```

```python
# SET CAV RANGE
session = zenoh.open(zenoh.Config.from_file('zenoh_config.json5'))
session_mgr = heros.zenoh.session_manager
session_mgr._session = session
with RemoteHERO('SCTL',session_manager=session_mgr) as Lock:
    Lock.update_setting('Cav', 'Master', 'range', [[0.5,0.7],[1.6,1.8]])
```

```python
# STOP VEXLUM LOCK
session = zenoh.open(zenoh.Config.from_file('zenoh_config.json5'))
session_mgr = heros.zenoh.session_manager
session_mgr._session = session
with RemoteHERO('SCTL',session_manager=session_mgr) as Lock:
    Lock.stop_loop('Lock1')
```
